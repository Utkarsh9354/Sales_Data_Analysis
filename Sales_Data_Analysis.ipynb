# Which is the best month for sale ?

```python
all_data['Order Date'][0].split('/')[0]
```

```python
def return_month(x):
    return x.split('/')[0]
```

```python
all_data['Month'] = all_data['Order Date'].apply(return_month)
```

```python
all_data['Month'].unique()
```

```python
filter = all_data['Month'] == 'Order Date'

all_data[~filter]
```

```python
# Excluding all those rows which have entry as "Order Date" in month feature .

all_data = all_data[~filter]
```

```python
# Convert 'Month', 'Quantity Ordered', and 'Price Each' columns to appropriate data types

all_data["Month"] = all_data["Month"].astype(int)
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype(int)
all_data['Price Each'] = all_data['Price Each'].astype(float)
```

```python
# creating sales feature

all_data['sales'] = all_data['Quantity Ordered'] * all_data['Price Each'] 
```

```python
all_data.groupby(['Month'])['sales'].sum()
```

```python
colors = ["#87CEEB", "#90EE90", "#FA8072", "#FFD700", "#F08080", "#FFB6C1", "#E0FFFF", "#E6E6FA", "#F5F5DC", "#D3D3D3", "#FFA07A", "#B0C4DE"]

# Plotting with the specified colors
ax = all_data.groupby(['Month'])['sales'].sum().plot(kind='bar', color=colors)

# Adding labels and title for better readability
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.set_title('Monthly Sales')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

# Which city has maximum order

```python
all_data.head()
```

```python
all_data['Purchase Address'][0]
```

```python
all_data['city'] = all_data['Purchase Address'].str.split(',').str.get(1)

all_data['city']
```

```python
total_orders = pd.value_counts(all_data['city'])

total_orders
```

```python
# Create a Pie Chart

explode = (0, 0, 0, 0, 0, 0, 0, 0, 0.15)

fig , ax = plt.subplots()
ax.pie(total_orders[::-1],
      autopct="%.0f%%",
      labels =  list(total_orders.index[::-1]),
      startangle = 90,
        explode=explode
      )

fig.suptitle("Share of Orders by City")
```

* Inference : New York , Los Angeles , San Francisco are the Top 3 cities which has max order 

# What product sold the most &  Why?

```python
all_data.head()
```

```python
count_df = all_data.groupby(['Product']).agg({'Quantity Ordered':'sum' , 'Price Each':'mean'}).reset_index()

count_df
```

```python
fig , ax = plt.subplots()

ax2 = ax.twinx()

x = count_df["Product"]

ax.bar(x ,
        count_df['Quantity Ordered'],
        color='orange',
        label= "Total Quantity Ordered",
       alpha = 0.45)

ax.set_ylabel('Order Count')
ax.set_xlabel("Products")
ax.legend(bbox_to_anchor=(1.6, 1), loc='upper right')
ax.set_xticklabels(x , rotation= 90 , fontsize=8 )
ax.set_title("Comparison of Product Prices and Quantity Ordered")

ax2.plot(x,
       count_df["Price Each"],
       label = "Median Price")

ax2.set_ylabel('Avg price of product')
ax2.legend(bbox_to_anchor=(1.2, 0.9), loc='upper left')

plt.show()
```

### Insights : 

* The top selling product is 'AAA Batteries'. 
* The top selling products seems to have a correlation with the price of the product. 
* The cheaper the product higher the quantity ordered and vice versa.

# What products are most often sold together ?

```python
all_data['Order ID'] 
```

```python
df_duplicated = all_data[all_data['Order ID'].duplicated(keep=False)]

df_duplicated
```

```python
duplicate_product = df_duplicated.groupby(['Order ID'])['Product'].apply(lambda x : ','.join(x)).reset_index().rename(columns={'Product':'grouped_products'})

duplicate_product.head()
```

```python
#Joining both the dataframes to get grouped data.

duplicate_product_df = df_duplicated.merge(duplicate_product , how='left' , on='Order ID')

duplicate_product_df
```

```python
no_duplicates = duplicate_product.drop_duplicates(subset = ["Order ID"])

no_duplicates
```

```python
#Top 5 grouped products sold

top_5 = no_duplicates['grouped_products'].value_counts()[0:5]

top_5.plot.pie(
    autopct='%0.f%%',
    startangle=90,
    colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']
)

plt.title('Top 5 Product Combinations Sold Together', fontsize=14, fontweight='bold')
plt.ylabel('')


plt.show()
```
